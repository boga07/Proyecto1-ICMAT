```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# ==========================================
# PARÁMETROS DEL PROYECTO (Fase 3)
# ==========================================
r_n = 3.91  # Tasa raza Nativa (Foshan)
r_i = 4.38  # Tasa raza Invasora (Tapachula)
N0 = 10     # Población inicial

# ==========================================
# CONFIGURACIÓN DINÁMICA
# ==========================================
try:
    K_input = float(input("Defina la población tope (L): "))
    t_max = float(input("Defina el rango de tiempo (Generaciones): "))
except ValueError:
    K_input = 5000.0
    t_max = 2.2

t = np.linspace(0, t_max, 500)

# ==========================================
# IMPLEMENTACIÓN DE MODELOS (Fase 1: Base)
# ==========================================
def malthus_continuo(N0, r, t):
    return N0 * np.exp(r * t)

def logistico_continuo(N0, r, K, t):
    return K / (1 + ((K - N0) / N0) * np.exp(-r * t))

# Cálculo de datos
m_n = malthus_continuo(N0, r_n, t)
m_i = malthus_continuo(N0, r_i, t)
l_n = logistico_continuo(N0, r_n, K_input, t)
l_i = logistico_continuo(N0, r_i, K_input, t)

# ==========================================
# CREACIÓN DEL GRÁFICO INTERACTIVO (PLOTLY)
# ==========================================
fig = go.Figure()

# Modelo de Malthus (Líneas sólidas: Celeste y Negro)
fig.add_trace(go.Scatter(x=t, y=m_n, mode='lines', name='Malthus: Nativa',
                         line=dict(color='skyblue', width=3)))
fig.add_trace(go.Scatter(x=t, y=m_i, mode='lines', name='Malthus: Invasora',
                         line=dict(color='black', width=3)))

# Modelo Logístico (Líneas sólidas: Azul y Rojo)
fig.add_trace(go.Scatter(x=t, y=l_n, mode='lines', name='Logístico: Nativa',
                         line=dict(color='blue', width=3)))
fig.add_trace(go.Scatter(x=t, y=l_i, mode='lines', name='Logístico: Invasora',
                         line=dict(color='red', width=3)))

# Línea de Capacidad de Carga (Punteada Verde)
fig.add_trace(go.Scatter(x=[0, t_max], y=[K_input, K_input], mode='lines',
                         name=f'Límite L={K_input}',
                         line=dict(color='green', width=2, dash='dash')))

# Configuración del diseño (Layout)
fig.update_layout(
    title='Dinámica Poblacional Interactiva: Aedes albopictus (A.3)',
    xaxis_title='Tiempo (Generaciones)',
    yaxis_title='Población (N)',
    yaxis=dict(range=[0, K_input * 1.3]),
    legend=dict(yanchor="bottom", y=0.01, xanchor="right", x=0.99), # Esquina inferior derecha
    hovermode="x unified", # Muestra todos los valores al mismo tiempo en el eje X
    template="plotly_white"
)

fig.show()
```

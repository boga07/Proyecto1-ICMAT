```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# PARÁMETROS DEL MODELO (Fase 3: Datos Reales)
# ==========================================
# Basado en la diferencia de fecundidad (50 vs 80 huevos)
# y velocidad de desarrollo de Aedes albopictus.

r_n = 3.91  # Tasa raza Nativa
r_i = 4.38  # Tasa raza Invasora

N0 = 10     # Población inicial
t = np.linspace(0, 2.5, 300) # Tiempo en generaciones

# ==========================================
# MODELO DE MALTHUS CONTINUO
# ==========================================
def malthus_continuo(N0, r, t):
    # Ecuación exacta: X(t) = X0 * e^(rt)
    return N0 * np.exp(r * t)

# Cálculo de las poblaciones
p_nativa = malthus_continuo(N0, r_n, t)
p_invasora = malthus_continuo(N0, r_i, t)

# ==========================================
# VISUALIZACIÓN EN ESCALA LINEAL
# ==========================================
plt.figure(figsize=(10, 6))

# Graficamos con escala normal para ver la curva exponencial
plt.plot(t, p_nativa, label='Raza Nativa (Foshan)', color='blue', linestyle='--')
plt.plot(t, p_invasora, label='Raza Invasora (Tapachula)', color='red', linewidth=2)

# Configuración de la gráfica
plt.title('Crecimiento Exponencial de Mosquitos (Modelo de Malthus)', fontsize=14)
plt.xlabel('Tiempo (Generaciones)', fontsize=12)
plt.ylabel('Cantidad de Mosquitos (N)', fontsize=12)

# Añadimos cuadrícula y leyenda
plt.grid(True, alpha=0.3)
plt.legend()

# Mostrar la gráfica
plt.show()
```
